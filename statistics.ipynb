```python
#Задание 2. Работа с выбросами¶
'''В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты.
Принять и обосновать решение о дальнейшей работе с ними.'''
```

```python
def outliers (param):
    q1 = horse_part[param].quantile(0.25)
    q3 = horse_part[param].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = horse_part[horse_part[param].between(lower_bound, upper_bound, inclusive=True)]
    print(pd.concat([horse_part, remove_outliers]).drop_duplicates(keep=False))
    
```

```python
#Задание 3. Работа с пропусками¶
'''Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками 
по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.'''
for col in horse_part.columns:
    pct_missing = horse_part[col].isna().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
horse_part.surgery.fillna(horse_part.surgery.median(), inplace=True) # заменяем медианой, поскольку это целое значение должго быть
horse_part.surgery.fillna(horse_part.surgery.median()).isna().sum()
```

```python
horse_part.pulse.fillna(horse_part.pulse.median(), inplace=True) # заменяем медианой, поскольку есть выбросы
horse_part.pulse.fillna(horse_part.pulse.median()).isna().sum()
```

```python
horse_part['abdominal distension'].fillna(horse_part['abdominal distension'].median(), inplace=True) # заменяем медианой, поскольку нужно целое число
horse_part['abdominal distension'].fillna(horse_part['abdominal distension'].median()).isna().sum()
```
