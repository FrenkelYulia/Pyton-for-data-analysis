```python
#Задание 1
'''Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class'''
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
def classifaer(param):
    if param <=2:
        return "низкий рейтинг"
    elif param <=4:
        return "средний рейтинг"
    else:
        return "высокий рейтинг"
    
    
ratings['class'] = ratings['rating'].apply(classifaer)
ratings
```

```python
#Задание 2
'''Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному
региону. Т. е. если поисковый запрос содержит название города региона, то в столбце 'region' пишется название этого региона.
Если поисковый запрос не содержит названия города, то ставим 'undefined'.
Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.'''
```

```python
keywords= pd.read_csv('keywords.csv')
keywords.head()
keywords.tail()
```

```python
def word_analis(keyword):
    centre =r'\b\москв[а-я]|\b\тул[а-я]|\b\ярославл[а-я]'
    nw = r'\b\петербург[а-я]|\b\псков[a-я]|\b\мурманск[а-я]'
    east =r'\b\владивосток[а-я]\b\сахалин[а-я]\b\хабаровск[а-я]'
    if len(re.findall(centre, keyword))!=0:
        return 'Центр'
    elif len(re.findall(nw, keyword))!=0:
        return 'Северо-Запад'
    elif len(re.findall(east, keyword))!=0:
        return 'Дальний Восток'
    else:
        return 'undefined'
keywords['region'] = keywords['keyword'].apply(word_analis)
keywords.region.value_counts()
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
def wordanalis(keyword):
    # проходим по ключам (название региона) и значениям (список городов региона) словаря geo_data 
    for region, city_list in geo_data.items():
        for city in city_list: # проходим по каждому городу региона
        # если название города есть в поисковом запросе, то возвращаем название региона region
            if city in keyword:
                return region

# если ничего не нашлось, то возвращаем undefined
    return 'undefined'
keywords['region'] = keywords['keyword'].apply(wordanalis)
keywords.region.value_counts()
```

```python
#Задание 3
'''Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, 
используя файлы с рейтингами фильмов из прошлого домашнего занятия 
(файл movies.csv и ratings.csv из базы https://grouplens.org/datasets/movielens). 
Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.
При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:
В переменную years запишите список из всех годов с 1950 по 2010.
Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов
содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
1) для каждой строки пройдите по всем годам списка years
2) если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
3) если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
4) Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’
5) Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

P/S/ все файлы находятся в архиве «дополнительные файлы к лекции и дз»'''
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
def production_year(title):
    pattern = r'\d{4}'
    k = re.findall(pattern, title)
    k =''.join(k)
    return k
    
joined['year'] = joined['title'].apply(production_year)
```
